```python
import polars as pl
from edges import combine, branch_count, in_both_branches, resembling_mentions
```

```python
df = combine()
```

```python
df = df.with_columns(
        in_both_branches(branch_count()).alias('in_both')
    )
```

```python
df.filter(~pl.col('in_both') & pl.col('branch').eq('main'))
```

```python
df.filter(~pl.col('in_both') & pl.col('branch').eq('refactored'))
```

```python
df.filter(
    ~pl.col('in_both')
).filter(
    pl.col('branch').eq('refactored') & ~(pl.col('resembling') & pl.col('type').eq('mention'))
)
```

```python
df.filter(
    ~pl.col('in_both') & pl.col('branch').eq('refactored')
).group_by(['type', 'resembling']).len()
```

```python
df.filter(~pl.col('in_both') & pl.col('branch').eq('main') & pl.col('_to').str.contains('orphan'))
```

```python
df.filter(pl.col('_from').str.contains('kv14.8') & ~pl.col('in_both'))
```
